```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/drive/MyDrive/CognoRise Infotech/Machine Learning/data.csv')
df
```

```python
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
df.dtypes
```

```python
# changing output to full form

df['diagnosis']=df['diagnosis'].map({'M':'Malignant','B':'Benign'})
df.head()
```

```python
col=list(df.columns)
col=col[1:]
col
```

```python
# Outliers Removing

for i in col:
 plt.figure(figsize=(5, 5))
 sns.boxplot(df[i], color='red')
 plt.xlabel(i)
 plt.ylabel('Range')
```

```python
# outliers removed
for i in col:

  def iqr_method(df,i):
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    upper=q3+(1.5*iqr)
    lower=q1-(1.5*iqr)
    return lower,upper
  lower_lim,upper_lim=iqr_method(df,i)
  df[i]=np.where(df[i]>upper_lim,upper_lim,np.where(df[i]<lower_lim,lower_lim,df[i]))
```

```python
for i in col:
 plt.figure(figsize=(5, 5))
 sns.boxplot(df[i], color='green')
 plt.xlabel(i)
 plt.ylabel('Range')
```

```python
x=df.drop(['diagnosis'],axis=1).values
y=df['diagnosis'].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
scaler.transform(x_train)
scaler.transform(x_test)
x_train
```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
knn=KNeighborsClassifier()
tree=DecisionTreeClassifier()
forest=RandomForestClassifier()
lst=[knn,tree,forest]
```

```python
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('Model is : ', tree)
  print('*'*100)
  print('Output :')
  print(y_pred)
  print('*'*100)
  print('Confusion Matrix :')
  print(confusion_matrix(y_test,y_pred))
  print('*'*100)
  print('Accuracy Score =',accuracy_score(y_test,y_pred))
  print('Classification Report :')
  print(classification_report(y_test,y_pred))
  print('*'*100)
```
